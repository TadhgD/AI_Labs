<h1>Pandas</h1>
In Pandas, there are two data structures.

Pandas allows us to easily read in CSV files.

The first type of data structure is a series, which is a one dimensional data structure.

```python
s = pd.Series([2, 4, -12, 0, 2])
```

```python
print(s)
print("The datatype :")
print(s.dtype)
print("The shape :")
print(s.shape)
```

We can also do similar things like we can do in numpy, for exmaple .dtype, .shape

Can also do s.values, which turns the Series into an array!

```python
s.values #
```

```python
s[3] #How we index a Series in pandas
```

```python
s[1:3] #Giz a slice
```

```python
s * 2 #Can perform elementwise operations in Pandas also.
```

```python
s > 0 #Apply an operator to each value in the array.
```

```python
s[s > 0] #We get back those that are true out of s. In this case, those that are greater than 0.
```

```python
print(s.sum())
print(s.mean())
```

```python
s.unique() #Notice how '2' is missing. Only returns the unique values.
```
